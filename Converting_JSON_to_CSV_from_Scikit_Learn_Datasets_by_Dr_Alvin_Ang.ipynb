#1. Import and Load Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#look at the JSON format....
iris
```



---



#2. Extracting the Dataframe out from JSON

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df
```

Screenshot_1.jpg



---



#3. Using Reset + Set_Index to Rename the Index Column to "Row No."

```python
df = df.reset_index()
df
```

```python
df = df.rename(columns={"index": "Row No."})
df
```

```python
df = df.set_index('Row No.')
df
```



---



#4. Extracting out the "Iris Class"

```python
y = pd.DataFrame(data=iris.target, columns=["Iris Class"])
```

```python
y
```



---



#5. Using Map to Create a New Column Z
To Map the Iris Type


*   0: Setosa
*   1: Versicolor
*   2: Virginica



```python
z = y["Iris Class"].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})
z
```



---



#6. Combining to get the Final Table...

```python
df["Iris Type"] = z
```

```python
df
```
